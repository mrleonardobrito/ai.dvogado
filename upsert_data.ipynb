```python
from langchain_community.document_loaders import PyPDFLoader

loader = PyPDFLoader("data/CF88_Livro_EC91_2016.pdf")
pages = loader.load_and_split()
```

```python
from langchain_text_splitters import CharacterTextSplitter

text_splitter = CharacterTextSplitter(chunk_size=1000, chunk_overlap=0)
docs = text_splitter.split_documents(pages)
```

```python
import os
from getpass import getpass

os.environ["PINECONE_API_KEY"] = getpass("Use your pinecone API KEY: ")
os.environ["OPENAI_API_KEY"] = getpass("Use your OpenAI API KEY: ")
os.environ["VOYAGE_API_KEY"] = getpass("Use your VoyageAI API KEY: ")
```

```python
from langchain_voyageai import VoyageAIEmbeddings

embeddings = VoyageAIEmbeddings(model="voyage-2")
```

```python
from langchain_pinecone import PineconeVectorStore

index_name = "advogado-index"

docsearch = PineconeVectorStore.from_documents(docs, embeddings, index_name=index_name)
```
